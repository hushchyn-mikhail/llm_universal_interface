## Датасет [Diabets](https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database)

```python
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database") + "/diabetes.csv"
df = pd.read_csv(path)
df.head()
X = df.drop(columns=["Outcome"])
y = df["Outcome"]
le = LabelEncoder()
y = le.fit_transform(y)
categorical_features = []

res_diabets = run_experiment(X=X, y=y, categorical_features=categorical_features)
```

## Датасет [Heart](https://www.kaggle.com/datasets/fedesoriano/heart-failure-prediction)

```python
path = kagglehub.dataset_download("fedesoriano/heart-failure-prediction") + "/heart.csv"

df = pd.read_csv(path)
df.head()

X = df.drop(columns=["HeartDisease"])
y = df["HeartDisease"]
le = LabelEncoder()
y = le.fit_transform(y)
categorical_features = ["Sex", "ChestPainType", "FastingBS", "RestingECG", "ExerciseAngina", "ST_Slope"]

res_heart = run_experiment(X=X, y=y, categorical_features=categorical_features)
```
